A function to reove NA's and duplicates

```python
def clean_dataframe(df):
    """
    Remove rows with missing values and duplicate rows from a DataFrame.
    Returns a cleaned copy of the DataFrame.
    """
    df_clean = df.dropna()            # remove rows with any NA
    df_clean = df_clean.drop_duplicates()  # remove duplicate rows
    df_clean = df_clean.reset_index(drop=True)  # tidy index
    return df_clean
```

create a function that goes though every numerical col, removes utliers and outputs a clean DF and an outlier DF

```python
# Loadi in a dataset
import seaborn as sns
df = sns.load_dataset("titanic")
df.head()
```

```python
import pandas as pd

def remove_outliers(df):
    """
    Detects and removes outliers from all numerical columns using the IQR method.
    Returns:
        df_clean  - DataFrame without outliers
        df_outliers - DataFrame containing only outlier rows
    """
    df_numeric = df.select_dtypes(include='number')
    df_outliers = pd.DataFrame()

    # Start with all rows marked as non-outliers
    mask = pd.Series([True] * len(df), index=df.index)

    for col in df_numeric.columns:
        Q1 = df[col].quantile(0.25)
        Q3 = df[col].quantile(0.75)
        IQR = Q3 - Q1

        lower = Q1 - 1.5 * IQR
        upper = Q3 + 1.5 * IQR

        # Identify outliers for this column
        col_outliers = (df[col] < lower) | (df[col] > upper)

        # Update global mask
        mask = mask & (~col_outliers)

    # Clean DF = rows that are never outliers
    df_clean = df[mask].reset_index(drop=True)

    # Outlier DF = rows removed
    df_outliers = df[~mask].reset_index(drop=True)

    return df_clean, df_outliers
```
